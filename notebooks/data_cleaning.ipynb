```python
import pandas as pd
import numpy as np
import zipfile


with zipfile.ZipFile("../.vscode/data/raw/Accidents0515.csv (1).zip", 'r') as z:
    df_accidents = pd.read_csv(z.open("Accidents0515.csv"), low_memory=False)


with zipfile.ZipFile("../.vscode/data/raw/Casualties0515.csv.zip", 'r') as z:
    df_casualties = pd.read_csv(
        z.open("Casualties0515.csv"), 
        low_memory=False, 
        on_bad_lines='skip'  # Struktur xətası olan sətirləri keç (atla)
    )


with zipfile.ZipFile("../.vscode/data/raw/Vehicles0515.csv.zip", 'r') as z:
    df_vehicles = pd.read_csv(
        z.open("Vehicles0515.csv"), 
        low_memory=False, 
        on_bad_lines='skip'  # Hər ehtimala qarşı bura da əlavə edirik
    )

print(f" Qəzalar: {df_accidents.shape[0]} sətir")
print(f" Xəsarətlər: {df_casualties.shape[0]} sətir")
print(f" Avtomobillər: {df_vehicles.shape[0]} sətir")
```

```python
df_accidents['Date'] = pd.to_datetime(df_accidents['Date'], dayfirst=True)
print(df_accidents['Date'])
```

```python
df_accidents['Year'] = df_accidents['Date'].dt.year
print(df_accidents['Year'])
```

```python
df_accidents['Month'] = df_accidents['Date'].dt.month
print(df_accidents['Month'])
```

```python
df_accidents['DayOfWeek'] = df_accidents['Date'].dt.dayofweek
print(df_accidents['DayOfWeek'])
```

```python
print(df_casualties.dtypes)
```

```python
casualty_cols_with_minus1 = ['Pedestrian_Location','Pedestrian_Movement','Car_Passenger','Bus_or_Coach_Passenger','Pedestrian_Road_Maintenance_Worker','Casualty_Home_Area_Type']

for col in casualty_cols_with_minus1:
    df_casualties[col] = df_casualties[col].replace(-1, np.nan)

df_casualties['Age_of_Casualty'] = df_casualties['Age_of_Casualty'].replace(-1, np.nan)

print("Əvəzləmədən sonra -1-lərin sayı:")
print((df_casualties[casualty_cols_with_minus1] == -1).sum())
```

```python
print(df_vehicles.dtypes)
```

```python
vehicle_cols_with_minus1 = ['Age_of_Driver','Age_of_Vehicle','Engine_Capacity_(CC)','Driver_IMD_Decile','Driver_Home_Area_Type','Journey_Purpose_of_Driver','Sex_of_Driver']

for col in vehicle_cols_with_minus1:
    df_vehicles[col] = df_vehicles[col].replace(-1, np.nan)

df_vehicles['Age_Band_of_Driver'] = df_vehicles['Age_Band_of_Driver'].replace(-1, np.nan)

print("Əvəzləmədən sonra -1-lərin sayı:")
print((df_vehicles[vehicle_cols_with_minus1] == -1).sum())
```

```python
for name, df in [('accidents', df_accidents),
                  ('casualties', df_casualties),
                  ('vehicles', df_vehicles)]:
    print(f"\n=== {name} ===")
    print(f"Shape: {df.shape}")
    nulls = df.isnull().sum()
    print(f"NaN olan sütunlar:\n{nulls[nulls > 0]}")
```
